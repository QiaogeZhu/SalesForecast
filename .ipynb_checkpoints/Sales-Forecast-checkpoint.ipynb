The project is expected to be complete and logical but not to be perfect and unbeatable.

# Data Preprocessing

In this step, we will deal with our features data, such as filling missing values, encoding categorical variables, merging dataset, normalizing features data into the same scales, etc.  

We will start with imputing and filling our missing values.

## Imputation of missing values  

In the previous section, we check that there are some missing values in features dataset.   
For five markdown variables, we fill the missing values with 0 since NaN indicates that there is no promotion on that day.  
For `CPI` and `Unemployment`, we will fill them with the median for each different store.  

```python
# markdown
markdown = {'MarkDown1': 0, 'MarkDown2': 0, 'MarkDown3': 0, 'MarkDown4': 0, 'MarkDown5': 0}
features_done = features.fillna(value=markdown)
```

```python
# CPI and Unemployment 
features_done['CPI'] = features_done.groupby('Store')['CPI'].transform((lambda x: x.fillna(x.median())))
features_done['Unemployment'] = features_done.groupby('Store')['Unemployment'].transform((lambda x: x.fillna(x.median())))
```

## Correct the `Type` values

```python
type_c_mean = round(stores.groupby('Type')['Size'].mean()['C'],2)
store_num = stores[(stores['Size']<type_c_mean) & ((stores['Type']=='A')|(stores['Type']=='B'))]['Store'].values
store_num
```

```python
stores_done = stores.copy()
stores_done.loc[stores['Store'].isin(store_num),'Type']='C'
```

## Data merging  

Before we handle the categorical variables, we need to merge dataset first since some of them have common categorical data.  
Merging dataset can extract more useful information during training and better predict our target variable.  

```python
train = pd.merge(train,features_done,on=['Store','Date','IsHoliday'])
train = pd.merge(train,stores_done,on=['Store'])

test = pd.merge(test,features_done,on=['Store','Date','IsHoliday'])
test = pd.merge(test,stores_done,on=['Store'])
```

```python
train.shape, test.shape
```

## Data encoding   

For some of models, categorical variables can not directly be trained, instead, they need to be transformed into appropriate format. Usually, we use label encoding for un-ordinal category, one-hot encoding for ordinal data, and 0/1 for binary variable.  

```python
# IsHoliday
train['IsHoliday'] = train['IsHoliday'].map({True: 1, False: 0})
test['IsHoliday'] = test['IsHoliday'].map({True: 1, False: 0})

# Type
train = pd.get_dummies(train, columns=['Type'])
test = pd.get_dummies(test, columns=['Type'])
```

```python
train.head(3)
```

```python
test.head(3)
```

## Feature Engineering 

At this step, we are going to extract more features from `Date`.  
Not the same day for every month or every year is the holiday. In other words, year, month, day might have some effects on stores' weekly sales.   
Therefore, we will extract `Year`, `Month`, `Day` from `Date`.  

```python
train['Year'] = pd.DatetimeIndex(train['Date']).year.astype(int)
train['Month'] = pd.DatetimeIndex(train['Date']).month.astype(int)
train['Day'] = pd.DatetimeIndex(train['Date']).day.astype(int)

test['Year'] = pd.DatetimeIndex(test['Date']).year.astype(int)
test['Month'] = pd.DatetimeIndex(test['Date']).month.astype(int)
test['Day'] = pd.DatetimeIndex(test['Date']).day.astype(int)
```

```python
train = train.drop('Date',axis=1)
test = test.drop('Date',axis=1)
```

```python
train.shape, test.shape
```

- `Fuel_Price` and `Year` are highly correlated.

```python
# dropping one of them is enough
drop_list = ['Fuel_Price','MarkDown1','Unemployment','CPI']
train = train.drop(drop_list, axis = 1)
```

```python
test = test.drop(drop_list, axis = 1)
```

# Modeling

Now, our data is ready for modeling. Since there is no target variable in our testing data, we need to create a validation subset to check the performance of our models. Therefore, 20% of training data will be the validation data.  

The scale of data will highly affect the performance of our models. It is highly recommended to scale our data before going into any further. Even PCA will require data scaling in advance. In this project, we will use Min Max Scaling to range the data from 0 to 1. This scaler perserves the shape of the original distribution, which in other words, does not change the information embedded in our original data. 

Also, we will create two kind of data, one with PCA and one without PCA. The reason why we are doing this is that PCA performs well when there is linear relationship within data, however, it will worsen our results if there is non-linear relationship within data.

Multiple machine learning models will be used to predict our target variable.  
- Linear Regression   
- Lasso
- Random Forest   
- Gradient Boosing Regression  
- Light GBM

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LassoCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
import lightgbm as lgb
```

### Data split

```python
x_train, x_val, y_train, y_val = train_test_split(train.drop('Weekly_Sales',axis=1), 
                                                  train['Weekly_Sales'], 
                                                  test_size=0.2, random_state=42)
x_test = test.copy()
x_train.shape, x_val.shape, x_test.shape
```

### Data scale

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_val_scaled = scaler.transform(x_val)
x_test_scaled = scaler.transform(x_test)
```

### PCA

```python
pca = PCA(0.95) # use the number of components that account for 95% of the variance
pca.fit(x_train_scaled)
x_train_pca = pca.transform(x_train_scaled)
x_val_pca = pca.transform(x_val_scaled)
x_test_pca = pca.transform(x_test_scaled)
```

```python
x_train_pca.shape, x_val_pca.shape, x_test_pca.shape
```

## Models 

### Linear Regression

Linear regression model may not be good enough to interpret the high-dimensional data, but we would like to have it as a first try.

```python
lr = LinearRegression().fit(x_train_scaled, y_train)
lr_val_pred = lr.predict(x_val_scaled)
```

```python
# pca
lr_pca = LinearRegression().fit(x_train_pca, y_train)
lr_pca_val_pred = lr_pca.predict(x_val_pca)
```

### Lasso   

Since Lasso has the characteristics to zero out least important features, PCA will not be applied here.

```python
lasso = LassoCV(cv=5, random_state=42)
lasso.fit(x_train_scaled, y_train)
lasso_val_pred = lasso.predict(x_val_scaled)
```

### Random Forest Regression

```python
rf = RandomForestRegressor(random_state = 42)
rf.fit(x_train_scaled, y_train)
rf_val_pred = rf.predict(x_val_scaled)
```

```python
# pca
rf_pca = RandomForestRegressor(random_state = 42)
rf_pca.fit(x_train_pca, y_train)
rf_pca_val_pred = rf_pca.predict(x_val_pca)
```

### Gradient Boosting Regression

```python
gb = GradientBoostingRegressor(random_state = 42)
gb.fit(x_train_scaled, y_train)
gb_val_pred = gb.predict(x_val_scaled)
```

```python
#with PCA
gb_pca = GradientBoostingRegressor(random_state = 42)
gb_pca.fit(x_train_pca, y_train)
gb_pca_val_pred = gb_pca.predict(x_val_pca)
```

### Light GBM  

```python
# create dataset for lightgbm 
lgb_train = lgb.Dataset(x_train_scaled, y_train.ravel())
lgb_val = lgb.Dataset(x_val_scaled, y_val.ravel(), reference=lgb_train)

# specify your configurations as a dict
params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': {'rmse','mae'},
    'num_leaves': 50,
    'max_depth': 6,
    'learning_rate': 0.01,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0
}

print('Starting training...')
# train the model
gbm = lgb.train(params,
             lgb_train,
             num_boost_round=10000,
             valid_sets=lgb_val,
             early_stopping_rounds=20,
             verbose_eval=False)
print('Saving model...')
gbm.save_model('lightgbm_model.txt')
print('Starting predicting...')
# predict
lgbm_val_pred = gbm.predict(x_val_scaled, num_iteration=gbm.best_iteration)
print('End predicting!')
```

```python
# pca
lgb_train_pca = lgb.Dataset(x_train_pca, y_train)
lgb_val_pca = lgb.Dataset(x_val_pca, y_val, reference=lgb_train_pca)

print('Starting training...')
# train the model
gbm_pca = lgb.train(params,
                lgb_train_pca,
                num_boost_round=10000,
                valid_sets=lgb_val_pca,
                early_stopping_rounds=20,
                verbose_eval=False)
print('Saving model...')
gbm_pca.save_model('pca_lightgbm_model.txt')
print('Starting predicting...')
# predict
lgbm_pca_val_pred = gbm_pca.predict(x_val_pca, num_iteration=gbm_pca.best_iteration)
print('End predicting!')
```

# Results/Conclusions  

Three kinds of metrics will be use here: $R^2$, RMSE, MAE.   
The prediction scores are made on the validation dataset rather than the real test dataset. 

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# show R2 score
# linear regression
r2_lr_val = r2_score(y_val,lr_val_pred)
r2_lr_pca_val = r2_score(y_val,lr_pca_val_pred)
# lasso
r2_lasso_val = r2_score(y_val,lasso_val_pred)
# random forests
r2_rf_val = r2_score(y_val,rf_val_pred)
r2_rf_pca_val = r2_score(y_val,rf_pca_val_pred)
# gradient boosting 
r2_gb_val = r2_score(y_val,gb_val_pred)
r2_gb_pca_val = r2_score(y_val,gb_pca_val_pred)
# light gbm
r2_lgbm_val = r2_score(y_val,lgbm_val_pred)
r2_lgbm_pca_val = r2_score(y_val,lgbm_pca_val_pred)

print("R2 Scores")
print("Linear Regression: ",r2_lr_val)
print("Linear Regression with PCA: ",r2_lr_pca_val)
print("Lasso: ",r2_lasso_val)
print("Random Forest: ",r2_rf_val)
print("Random Forest with PCA: ",r2_rf_pca_val)
print("Gradient Boosting: ",r2_gb_val)
print("Gradient Boosting with PCA: ",r2_gb_pca_val)
print("Light GBM: ",r2_lgbm_val)
print("Light GBM with PCA: ",r2_lgbm_pca_val)
```

```python
# show rmse score
# linear regression
mse_lr_val = mean_squared_error(y_val,lr_val_pred)
mse_lr_pca_val = mean_squared_error(y_val,lr_pca_val_pred)
# lasso
mse_lasso_val = mean_squared_error(y_val,lasso_val_pred)
# random forests
mse_rf_val = mean_squared_error(y_val,rf_val_pred)
mse_rf_pca_val = mean_squared_error(y_val,rf_pca_val_pred)
# gradient boosting
mse_gb_val = mean_squared_error(y_val,gb_val_pred)
mse_gb_pca_val = mean_squared_error(y_val,gb_pca_val_pred)
# light gbm
mse_lgbm_val = mean_squared_error(y_val,lgbm_val_pred)
mse_lgbm_pca_val = mean_squared_error(y_val,lgbm_pca_val_pred)

print("RMSE Scores")
print("Linear Regression: ",mse_lr_val**0.5)
print("Linear Regression with PCA: ",mse_lr_pca_val**0.5)
print("Lasso: ",mse_lasso_val**0.5)
print("Random Forest: ",mse_rf_val**0.5)
print("Random Forest with PCA: ",mse_rf_pca_val**0.5)
print("Gradient Boosting: ",mse_gb_val**0.5)
print("Gradient Boosting with PCA: ",mse_gb_pca_val**0.5)
print("Light GBM: ",mse_lgbm_val**0.5)
print("Light GBM with PCA: ",mse_lgbm_pca_val**0.5)
```

```python
# show mean absolute error
# linear regression
mae_lr_val = mean_absolute_error(y_val,lr_val_pred)
mae_lr_pca_val = mean_absolute_error(y_val,lr_pca_val_pred)
# lasso
mae_lasso_val = mean_absolute_error(y_val,lasso_val_pred)
# random forests
mae_rf_val = mean_absolute_error(y_val,rf_val_pred)
mae_rf_pca_val = mean_absolute_error(y_val,rf_pca_val_pred)
# gradient boosting
mae_gb_val = mean_absolute_error(y_val,gb_val_pred)
mae_gb_pca_val = mean_absolute_error(y_val,gb_pca_val_pred)
# light gbm
mae_lgbm_val = mean_absolute_error(y_val,lgbm_val_pred)
mae_lgbm_pca_val = mean_absolute_error(y_val,lgbm_pca_val_pred)

print("MAE Scores")
print("Linear Regression: ",mae_lr_val)
print("Linear Regression with PCA: ",mae_lr_pca_val)
print("Lasso: ",mae_lasso_val)
print("Random Forest: ",mae_rf_val)
print("Random Forest with PCA: ",mae_rf_pca_val)
print("Gradient Boosting: ",mae_gb_val)
print("Gradient Boosting with PCA: ",mae_gb_pca_val)
print("Light GBM: ",mae_lgbm_val)
print("Light GBM with PCA: ",mae_lgbm_pca_val)
```

### Conclusion

PCA does not perform very well as we expected. Compared with PCA, features selected by random forests algorithm work better. It might be because there are some nonlinear relationships between our features.    

Comparing with multiple models, we can see all three kinds of scores together. Random forest outputs the best result among others. 

### Limitations/Improvements

- Since the dataset is large and some of algorithms are computational expensive without the support of GPU, we did not try as many machine learning algorithms as we can.   


- Grid search will be a good tool to look for a better combination of parameters or hyperparameters used in our models, however, for the same reason that the limitation of the running time, we gave up using this useful tool.    


- When exploring the dataset, we did not check any outliers. It is highly possible that some of our models would get improved if we leave out any possible outliers. However, outliers do not influence that much when training with Random forest, our final pick might not change.   


- Our results are not good compared to competitors who submitted their results on Kaggle (which untilized weighted mean absolute error as the metric with more weight on IsHoliday=True and less weight on IsHoliday=False), however, we tried our best and paid much focus on what we have learned so far.
